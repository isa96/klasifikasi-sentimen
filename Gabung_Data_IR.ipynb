```python
import pandas as pd
```

```python
var = ['date','tweet']
df1 = pd.read_csv("thomascup_090522.csv", names=var)
df2 = pd.read_csv("thomascup_100522.csv", names=var)
df3 = pd.read_csv("thomascup_110522.csv", names=var)
df4 = pd.read_csv("thomascup_120522.csv", names=var)
df5 = pd.read_csv("thomascup_130522.csv", names=var)
df6 = pd.read_csv("thomascup_140522.csv", names=var)
df7 = pd.read_csv("thomascup_150522.csv", names=var)
df8 = pd.read_csv("thomascup_160522.csv", names=var)
df9 = pd.read_csv("thomascup_170522.csv", names=var)
```

```python
df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9]
```

```python
df_list
```

```python
gabungan = pd.concat(df_list, ignore_index=True, axis=0)
```

```python
gabungan.head()
```

```python
#pilih data yang ingin digunakan saja
df  = pd.DataFrame(gabungan['tweet'])
df
```

```python
import pandas as pd
import numpy as np
import re
#remove user
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt    
df['remove_user'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
df.drop_duplicates(subset ="remove_user", keep = 'first', inplace = True)
df.sort_values("remove_user", inplace = True)
```

```python
df.head()
```

```python
data_labeling = df['remove_user']
data_labeling.to_csv("thomascup_gabungan.csv",encoding='utf-8', index=False)
```
